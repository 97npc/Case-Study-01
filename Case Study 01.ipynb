## 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
count=data["Buyer Gender"].value_counts()
```

```python
count
```

```python
data["Buyer Gender"].value_counts().plot.bar(width=0.6,color=["red","blue"])
plt.xlabel("Buyer Gender")
plt.ylabel("Count")
plt.title("Male vs Female Buyers")
plt.grid()
plt.show()
```

DATA INSIGHT

1.The total no. of buyers including men and women are 10000.

2.Out of total numbers, there are 5052 female buyers and 4948 male buyers.

3.The data provides that there are more female buyers than the male buyers.

## 4. Find the top 5 cars based on their sales price.

```python
#sort sales price in descending order
data=data.sort_values("Sale Price",ascending=False)
```

```python
Top_5_cars=data[["Make","Sale Price"]]
```

```python
Top_5_cars.reset_index(drop=True).head()
```

## 5. Find the least 5 cars based on their Resell price.

```python
#sort resell price in ascending order
data=data.sort_values("Resell Price",ascending=True)
```

```python
Least_5_cars=data[["Make","Resell Price"]]
```
